__Decision tree uses a tree-like model of decisions.It is used both classification and regression for predicting a data.A decision tree can be used to visually and explicitly represent decisions and decision making.__

```python
#import labrary
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split as tts
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,f1_score
from sklearn import tree
from matplotlib import pyplot as plt
from sklearn import preprocessing
```

## Load Data

```python
iris=pd.read_csv('Iris (1).csv')
iris.head()
```

```python
#imformation about load data
iris.info()
```

```python
iris.isnull().sum()
```

```python
x=iris.values[:,1:5]
y=iris.values[:,5]
print(y)
```

## Encode target data

```python
le=preprocessing.LabelEncoder()
y=le.fit_transform(y)
y
```

## Split data into Train and Test  

```python
#split data
X_train,X_test,y_train,y_test=tts(x,y,test_size=0.2,random_state=0)
```

```python
dtree=DecisionTreeClassifier(criterion='entropy',max_depth=5)
dtree.fit(X_train,y_train)
```

```python
y_pred=dtree.predict(X_test)
```

## Confusion Matrix

```python
print(confusion_matrix(y_test,y_pred))
```

## Check accuracy of train model

```python
print(accuracy_score(y_pred,y_test))
```

## F1-Score

```python
print(f1_score(y_test,y_pred,average='weighted'))
```

# Visualization of Decision Tree

```python
t=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(dtree, 
                   feature_names=t,  
                   class_names=iris['Species'],
                   filled=True,rounded=True)
```
